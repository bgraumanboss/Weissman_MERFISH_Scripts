## Merge probes

```python
probe_filenames = [mf4_16bit_probe_filename, mf4_16bit_dna_probe_filename, mf5_probe_filename]

twist_probe_df = {
    'name':[],
    'sequence':[],
    'library':[],
}
lib_names = []
for _pb_file in probe_filenames:
    _probe_df = pd.read_excel(_pb_file)
    _lib_name = _pb_file.split(os.path.sep)[-2].replace('_','-')
    print(_lib_name)
    lib_names.append(_lib_name)
    for _ipb, _pb in _probe_df.iterrows():
        twist_probe_df['name'].append(_lib_name+"_"+_pb['name'])
        twist_probe_df['sequence'].append(_pb['sequence'])
        twist_probe_df['library'].append(_lib_name)
# convert to df
twist_probe_df = pd.DataFrame(twist_probe_df)
print(len(twist_probe_df))
```

```python
twist_probe_df
```

## Plot summaries

```python
library_2_info = {}
for _lib in lib_names:
    _lib_df = twist_probe_df.loc[twist_probe_df['library']==_lib]
    
    # generate info dataframe:
    _lib_info = {
        'gene':[],
        'probeNum': [],
        'readouts': [],
    }
    for _i, _pb in _lib_df.iterrows():
        _gene = _pb['name'].split('_')[1]
        _readouts = [_rd for _rd in _pb['name'].split('_')[-1].split(':') if len(_rd)]
        if _gene not in _lib_info['gene']:
            _lib_info['gene'].append(_gene)
            _lib_info['probeNum'].append(1)
            _lib_info['readouts'].append(_readouts)
        else:
            _gene_index = _lib_info['gene'].index(_gene)
            _lib_info['probeNum'][_gene_index] += 1
            for _rd in _readouts:
                if _rd not in _lib_info['readouts'][_gene_index]:
                    _lib_info['readouts'][_gene_index].append(_rd)
    # convert into df
    library_2_info[_lib] = pd.DataFrame(_lib_info)
    print(_lib, len(_lib_df), len(library_2_info[_lib]))
```

```python
save_folder = mf5_library_folder

twist_probe_filename = os.path.join(save_folder, 'mf4-mf5_probe.xlsx')
print(twist_probe_filename)
twist_probe_df[['name','sequence']].to_excel(twist_probe_filename, index=None)

twist_info_filename = os.path.join(save_folder, 'mf4-mf5_design_info.xlsx')
print(twist_info_filename)
# loop through libraries:
writer = pd.ExcelWriter(twist_info_filename, engine = 'xlsxwriter')
for _lib, _df in library_2_info.items():
    _df.to_excel(writer, index=None, sheet_name=_lib)
writer.close()
```
